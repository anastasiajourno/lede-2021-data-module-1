## 1. Merge

Your goal is to create a table that lists all committees with their associated candidate. Note: candidates usually have more than one committee.

1. Senate candidates seeking election in 2022 | [column definitions](https://www.fec.gov/campaign-finance-data/candidate-summary-file-description/)
1. A list of committees registered with the FEC | [column definitions](https://www.fec.gov/campaign-finance-data/committee-summary-file-description/)

```python
# load in both tables
candidates = pd.read_csv('downloaded_data/candidate_summary_2022.csv')
```

```python
committees = pd.read_csv('downloaded_data/committee_summary_2022.csv')
```

### a. What is the common attribute between these two tables?
Hint: look at the column definitions for each instead of guessing by their name

1. Committee: CAND_ID
1. Candidates: Cand_Id

### b. Fill in the following to determine which arguments/kwargs you would need to match candidates with all of their committees with pandas merge function?
- Left table: committees
- Right table: candidates
- Merge type: left
- Left ID: CAND_ID
- Right ID: Cand_Id

```python
committees_candidates = committees.merge(candidates, left_on='CAND_ID', right_on='Cand_Id', how='left')
```

```python
committees_candidates.head()
```

```python
# check the length of each table. confirm they are the length we would expect
len(candidates), len(committees), len(committees_candidates)
```

With this left join, which values were removed from the final table?
- committees with no candidate ID associated with it
- committees where the candidate ID listed is not in the candidates table

### c. What if we want to create a table with a candidate and their principal campaign committee?
Even though a candidate can have more than one committee affiliated with their campaign, they can only have one "principal" committee. What do we need to know to create this new table?

- Left table: candidates
- Right table: filtered committees table
- Merge type: inner
- Left ID: CAND_ID
- Right ID: Cand_Id

```python
# what additional filtering do we need to do and to which table?
principal_committees = committees[committees['CMTE_DSGN'] == 'P']
```

```python
# check the number of records
len(principal_committees)
```

```python
candidates_principal_committees = candidates.merge(principal_committees, left_on='Cand_Id', right_on='CAND_ID', how='inner')
```

```python
len(candidates), len(principal_committees), len(candidates_principal_committees)
```

With this inner join, which values were removed from the final table?
- committees with no candidate ID associated with it
- committees where the candidate ID listed is not in the candidates table
- candidates where the candidate ID is not in the committees table
